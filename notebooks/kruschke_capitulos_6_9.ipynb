**Distribucion Beta**

$$beta(\theta|a, b)=\frac{\theta^{a-1}(1-\theta)^{b-1}}{B(a, b)}$$

$$B(a, b)=\int_0^1{\theta^{a-1}(1-\theta)^{b-1}d\theta}$$
La distribucion beta es *dbeta(theta, a, b)* y la funcion beta es beta(a, b).


```r
x = c(1:100)/100
a = 0.1
b = 0.1
plot(x, dbeta(x, a, b)/beta(a, b))
```

#### Distribución Gamma

Dos parámetros: de forma, $\alpha$, y de escala inversa o tasa, $\beta = 1/\theta$, y ambos son positivos reales.

$$f(x;\alpha, \beta)=\frac{\beta^\alpha x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)}$$ for  $$x \gt 0$$ and $$\alpha, \beta \gt 0$$,

```r
x = c(1:100)/10
shape = 1
rate = 0.5
scale = 1/rate
print(paste('Shape:', shape, '   Scale:', scale, '   Rate:', rate, sep=''))
plot(x, dgamma(x, shape, rate))
```

```r
shape = 5
rate = 1
scale = 1/rate
print(paste('Shape:', shape, '   Scale:', scale, '   Rate:', rate, sep=''))
plot(x, dgamma(x, shape, rate))
```

Media 1, desviacion estandar grande y moda en cero:

```r
# Create the function mode
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}
```

```r
x = c(1:1500)/10
shape = 0.01
rate = 0.01
scale = 1/rate
y = round(dgamma(x, shape, rate), 4)
print(paste('Shape:', shape, '   Scale:', scale, '   Rate:', rate, sep=''))
print(paste('Mean:', mean(y), '   SD:', sd(y), '   Mode:', getmode(y), sep=''))
plot(x, y)
```
